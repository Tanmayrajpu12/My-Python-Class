```python
tmp = pd.pivot_table(df, values='Episode Number', columns='Season Number', aggfunc='max')
print(tmp)
```

```python
# Season-wise number of episodes - India SharkTank

x = tmp.plot.bar(figsize=(5,4), title="Season-wise number of episodes - India SharkTank")
plt.xlabel("Seasons")
plt.ylabel(' '.join("Number of episodes".upper()))
plt.xticks([])
for t in x.patches:
    if (np.isnan(float(t.get_height()))):
        x.annotate(0, (t.get_x(), 0))
    else:
        x.annotate(str(format(int(t.get_height()), 'd')), (t.get_x(), t.get_height()*1.01))
```
